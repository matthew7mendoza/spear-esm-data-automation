Make a default plot of your dataarray and print out the data contents

Don't forget to import xarray and numpy!

````python
import pandas as pd
import numpy as np
import xarray as xr
two_years = pd.date_range(start='2014-01-01', end='2016-01-01', freq='D')
fake_data = np.sin(2 *np.pi *two_years.dayofyear / 365)

# Setting everything up
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
%whos

# In the ipynb file, the % is the escape sequence
````

````python
da = xr.DataArray(fake_data, dims=['time'], coords = {'time':two_years}, attrs={'units':'whoopty','time':'years'})
da

# One dimension DataArray, with the only dimension being time, the coordinate on time is time which follows the data of two_years
# two_years defined by the pandas method. 
````

````python
da.plot(linestyle='--');
````
