```python
%load_ext autoreload
%autoreload 2

import sys
import os

sys.path.append(os.path.abspath(".."))

from src.document_reader import extract_text
from src.ai_agent import LLM_Agent
```

```python
agent = LLM_Agent(temperature = 0.0)
print("AI AGENT ACTIVATED")
```

```python
from pathlib import Path

TARGET_DOCUMENT_TYPE = "DMP"
FILE_PATH = Path("../test_metadata.txt")

def generate_document_draft(doc_type: str, file_path: Path) -> None:
    """
    Reads a file and uses the AI agent to extract and format a document draft.
    """

    try:
        print(f"1. Reading file: {file_path.name}")
        document_text = extract_text(file_path)
        print("File read successfully.\n")

        print(f"2. Processing {doc_type} extraction via LLM...")
        result = agent.process_document(doc_type, document_text)
        print("Extraction complete.\n")

        print(f"--------GENERATED {doc_type} DRAFT--------\n")

        for question, answer in result.get("extracted_answers", {}).items():
            print(f"{question}\n> {answer}\n")

        missing_info = result.get("missing_information", [])
        if missing_info:
            print("--------MISSING INFORMATION--------")
            print("The AI could not find the answers to the following questions:\n")
            for missing in missing_info:
                print(f"- {missing}")

    except FileNotFoundError:
        print(f"\nCould not locate file at {file_path}")
    except Exception as e:
        print(f"\nUnexpected error occured: {e}")

generate_document_draft(TARGET_DOCUMENT_TYPE, FILE_PATH)
```
